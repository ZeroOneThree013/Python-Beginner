```python
from PIL import Image
import numpy as np
import cv2

# 用 PIL 讀圖（支援中文路徑）
image_path = r'D:\金融科技\HW\圖檔\地球娛樂室.jpg'
pil_img = Image.open(image_path)

# 轉成 OpenCV 的 BGR 格式
image = cv2.cvtColor(np.array(pil_img), cv2.COLOR_RGB2BGR)

print(image.shape)
```

```python
plt.imshow(image)
plt.show()
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(
    gray, 
    scaleFactor = 1.1,
    minNeighbors = 5,
    minSize=(30,30),
    flags = cv2.CASCADE_SCALE_IMAGE
)

print(faces)
```

```python
import random
font = cv2.FONT_HERSHEY_SIMPLEX

for (x,y,w,h) in faces:
    cv2.rectangle(image,(x,y),(x+w,y+h),(14,201,255),2)
    cv2.putText(image, str(random.randrange(30,40)), (x+(w//2)-18,y-10), font, 1, (14,201,255), 2)
    cv2.rectangle(image, (image.shape[1]-140, image.shape[0]-20),
                  (image.shape[1], image.shape[0]), (0,255,255), -1)
    cv2.putText(image, 'Finding '+str(len(faces))+' face',
                (image.shape[1]-135, image.shape[0]-5),
                font, 0.5, (255,0,0), 1)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.show()

# cv2.imshow('Faces found',image)
# cv2.imwrite('Beatles.jpg',image)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```

```python
cv2.imwrite('Beatles.jpg',image)
num = 1
for (x,y,w,h) in faces:
    filename = 'face' + str(num) + '.jpg'
    image = Image.open('Beatles.jpg')
    image_crop = image.crop((x,y,x+w,y+h))
    # image_resize = image_crop.resize((150,150),Image.ANTIALIAS) # 高品質:ANTIALIAS
    image_resize = image_crop.resize((150,150), Image.Resampling.LANCZOS) # 新寫法
    image_resize.save(filename)
    num += 1
```
